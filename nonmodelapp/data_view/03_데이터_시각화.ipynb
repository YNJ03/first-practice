```python
plt.title("시각화 종합")

x = [10, 20, 30, 40]
y = [1, 4, 9, 16]

# c : 색상
# lw : 선의 굵기
# ls : 선 스타일
# marker : 지점 모양
# ms : 마커 사이즈
# mec : 마커 테두리 색상
# mew : 마커 테두리 두께
# mfc : 마커 내부 채우기

plt.plot(x, y, c="b", lw=5, ls='--', marker="o", ms=15, mec="g", mew=5, mfc="r")

## x축의 범위
plt.xlim(0, 50)

## y축의 범위
plt.ylim(-10, 30)

plt.show()
```

### seaborn 시각화(고급시각화에 속함)

```python
### 색상 변경하기
# 기본 팔레트 색상 : deep, muted, pastel, bright, dark, colorblind
sns.set_palette("deep")

sns.stripplot(x="day", y='total_bill', data=tips)
```

```python
### 막대 그래프
# - 오차막대 : 비교 데이터간의 차이를 표시함 (잘 안씀)
# - 오차막대를 안보이게 하려면 : estimator=len 추가
sns.barplot(x="sex", y='tip', data=tips, estimator=len)
```
