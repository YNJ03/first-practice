```python
"""
connect와 cursor는 매번 접속과 종료를 반복해야함
- 함수로 만들어서 사용..
- 자유롭게 함수이름 만든 후 사용..
- 생성한 함수를 이용해서 위에서 사용한 SQL구문을 사용하여 출력해보기
"""
```

```python
def getDataFrame(sql) :
    # 드라이브는 한번만 실행해도 됨(굳이 안넣어도 된다는 의미)
    dsn = cx_Oracle.makedsn('localhost', 1521, 'xe')
    # DB 접속
    conn = cx_Oracle.connect('busan', 'dbdb', dsn)
    # 커서 생성
    cursor = conn.cursor()
    # SQL 응답 받아오기
    rs = cursor.execute(sql)
    # 결과 rows에 담기
    rows = cursor.fetchall()
    
    # 컬럼명 가져와서 리스트로 만들기
    colnames = cursor.description
    cols_list = [colnames[i][0] for i in range(len(colnames))]
    
    # 데이터프레임 형태로 만들기
    df = pd.DataFrame(rows, columns = cols_list)
    
    # DB 접속 끊기
    cursor.close()
    conn.close()
    
    return df
```

```python
### 강사님 - 함수 만들기
def initConn() :
    conn = cx_Oracle.connect('busan', 'dbdb', dsn)
    cursor = conn.cursor()
    
    return conn, cursor

def getDataFrame2(conn, cursor) :
    rows = cursor.fetchall()

    colnames = cursor.description
    cols_list = [colnames[i][0] for i in range(len(colnames))]

    cursor.close()
    conn.close()

    return pd.DataFrame(rows, columns = cols_list)
```

```python
# - 두번째 방법
conn, cursor = initConn()

sql = """
        select *
        from member
        where mem_id = :mem_id
        and mem_mileage >= :mem_mileage
"""
cursor.execute(sql, mem_id='b001', mem_mileage = 1000)

df7 = getDataFrame2(conn, cursor)
df7
```

```python
"""
회원 거주지역별 주문수량의 합을 조회하기
- 정렬은 거주지역 오름차순
- 상품분류명에 "전자제품"이 포함되고
- 거래처명에 "대우전자"가 포함된 경우
함수는 편하신 것으로 사용
- 데이터프레임 변수명 : df8
"""
```

```python
# 전체 함수(첫번째 방법)
sql = """select substr(mem_add1,1,2) as area , sum(cart_qty) as sum_qty
        from member inner join cart
                    on (mem_id = cart_member)
                    inner join prod
                    on (cart_prod = prod_id)
                    inner join lprod
                    on (prod_lgu = lprod_gu and lprod_nm like '%{}%')
                    inner join buyer
                    on (prod_buyer = buyer_id and buyer_name like '%{}%')
        group by substr(mem_add1,1,2)
        order by area ASC

""".format('전자제품', '대우전자')

df8 = getDataFrame(sql)
df8
```

```python
# 분할 함수(두번째 방법)
conn, cursor = initConn()

sql = """
        select substr(mem_add1,1,2) as area , sum(cart_qty) as sum_qty
        from member inner join cart
                    on (mem_id = cart_member)
                    inner join prod
                    on (cart_prod = prod_id)
                    inner join lprod
                    on (prod_lgu = lprod_gu and lprod_nm like :lprod_nm)
                    inner join buyer
                    on (prod_buyer = buyer_id and buyer_name like :buyer_name)
        group by substr(mem_add1,1,2)
        order by area ASC
"""
cursor.execute(sql, lprod_nm='%전자제품%', buyer_name = '%대우전자%')

df8 = getDataFrame2(conn, cursor)
df8
```

```python
### 세번째 방법
# - ** : 여러 값을 알아서 받아서 표시함(딕셔너리 형태로)
def getDataFrame3(sql, **params) :
    print(params)
    conn = cx_Oracle.connect('busan', 'dbdb', dsn)
    cursor = conn.cursor()

    cursor.execute(sql, params)
    rows = cursor.fetchall()
    
    # 컬럼명 가져와서 리스트로 만들기
    colnames = cursor.description
    cols_list = [colnames[i][0] for i in range(len(colnames))]
    
    # 데이터프레임 형태로 만들기
    df = pd.DataFrame(rows, columns = cols_list)
    
    # DB 접속 끊기
    cursor.close()
    conn.close()
    
    return df
```
